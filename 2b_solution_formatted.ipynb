```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("2b.csv", sep=';')
```

```python
df = df.drop("Unnamed: 7", axis=1)
```

```python
df = df.drop("Grouping example (first 20 transactions)", axis=1)
```

```python
df = df.sort_values(by=["Maturity period, days", "Rate, %"], ascending = [True, True])
```

```python
df
```

```python
print(df["Rate, %"].min())
print(df["Rate, %"].max())

print(df["Maturity period, days"].min())
print(df["Maturity period, days"].max())
```

```python
gr_number = 1
```

```python
for i in range(8495, 91251, 1500):  # by interest rates
    for j in range(32, 93, 30):  # by maturity period
        df.loc[(df["Maturity period, days"] == j) & (df["Rate, %"] >= i / 10000) & (df["Rate, %"] < (i + 1500) / 10000), "ANSWER (GROUP NUMBER)"] = gr_number
        gr_number += 1
```

```python
df["ANSWER (GROUP NUMBER)"].unique()
```

```python
df = df.fillna(-1)
```

```python
df["ANSWER (GROUP NUMBER)"] = df["ANSWER (GROUP NUMBER)"].astype(int)
```

```python
groups = np.delete(df["ANSWER (GROUP NUMBER)"].unique(), -1)  # actual groups
```

```python
rwa = 0
```

```python
for val in df["ANSWER (GROUP NUMBER)"].unique():
    A = df.loc[df["ANSWER (GROUP NUMBER)"] == val, "Asset (active)"].sum()  # sum of the assets in a group
    L = df.loc[df["ANSWER (GROUP NUMBER)"] == val, "Loan (passive))"].sum()  # sum of the loans in a group
    if val in groups:
        if A != 0 and L != 0:
            rwa += abs(A - L)  # netting
            continue
        df.loc[df["ANSWER (GROUP NUMBER)"] == val, "ANSWER (GROUP NUMBER)"] = -1
    df.loc[df["ANSWER (GROUP NUMBER)"] == val, "ANSWER (GROUP NUMBER)"] = -1
```

```python
A_not_netted = sum(df.loc[df["ANSWER (GROUP NUMBER)"] == -1, "Asset (active)"])  # sum of the not netted assets
L_not_netted = sum(df.loc[df["ANSWER (GROUP NUMBER)"] == -1, "Loan (passive))"])  # sum of the not netted loans
```

```python
rwa += (A_not_netted + L_not_netted) * 0.1 + abs(A_not_netted - L_not_netted) * 0.4
```

```python
rwa  # answer, 10245291618.791994
```
