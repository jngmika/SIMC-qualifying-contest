```python
import pandas as pd
```

```python
df = pd.read_csv("2a.csv", sep=';')
```

```python
df = df.drop("Unnamed: 6", axis=1)
```

```python
df = df.sort_values("Rate, %")
```

```python
df
```

```python
gr_number = 1
for i in range(-500, 801, 15):
    df.loc[(df["Rate, %"] >= i / 100) & (df["Rate, %"] < (i + 15) / 100), "ANSWER (GROUP NUMBER)"] = gr_number
    gr_number += 1
```

```python
df.loc[df["Rate, %"] > 7.9, "ANSWER (GROUP NUMBER)"] = gr_number
```

```python
df["ANSWER (GROUP NUMBER)"] = df["ANSWER (GROUP NUMBER)"].astype(int)
```

```python
df #just to see it
```

```python
df["ANSWER (GROUP NUMBER)"].unique()
```

```python
rwa = 0
```

```python
for val in df["ANSWER (GROUP NUMBER)"].unique():
    A = df.loc[df["ANSWER (GROUP NUMBER)"] == val, "Asset (active)"].sum()  # sum of the assets in a group
    L = df.loc[df["ANSWER (GROUP NUMBER)"] == val, "Loan (passive))"].sum()  # sum of the loans in a group
    if A != 0 and L != 0:
        rwa += abs(A - L)
    else:
        A = df.loc[df["ANSWER (GROUP NUMBER)"] == val, "ANSWER (GROUP NUMBER)"] = -1
```

```python
A_not_netted = sum(df.loc[df["ANSWER (GROUP NUMBER)"] == -1, "Asset (active)"])  # sum of the not netted assets
L_not_netted = sum(df.loc[df["ANSWER (GROUP NUMBER)"] == -1, "Loan (passive))"])  # sum of the not netted loans
```

```python
rwa += (A_not_netted + L_not_netted) * 0.1 + abs(A_not_netted - L_not_netted) * 0.4
```
